# Question 2

Outlier: An outlier is bservation in a dataset that significantly deviates from the majority of the data points.

Importance of Outliers:
1. Outliers can indicate errors in data collection or entry.
2. They can skew summary statistics and lead to incorrect conclusions.
3. Outliers can also contain valuable information or indicate interesting phenomena.

# Question 3

Three common ways of detecting outliers in the given dataset:

a. Z-Score Method: This method is used to identify outliers in a dataset by measuring the amount of stadard deviations the data point is away from the mean. It works by: 

    i. Calculating the mean(μ) and standard deviation(σ) of the dataset.
    
    ii. For each data point, the Z-Score is computed using the formula: Z = (X - μ) / σ, where X is the data point.
    
    iii. Normally, a threshold value which is normally set to 3 or -3, is used. Data points with Z-Scores that exceed the threshold are considered outliers.
    

b. Interquartile Range (IQR) Method: This method is also used to identify outliers and is not affected by extreme values. It works by:

    i. Calculating the first quartile (Q1) and third quartile (Q3) of the dataset.
        
    ii. Computing the IQR, which is the range between Q3 and Q1: IQR = Q3 - Q1.
    
    iii. Data points below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR are considered outliers.
        
c. Data Visualization (Box Plot): This method involves displaying the the distribution of the dataset, showing the median, quartiles, and potential outliers. The outliers are often represented as individual points beyond "whiskers" extending from the boxes. The data points that fall outside the whiskers are considered outliers.
